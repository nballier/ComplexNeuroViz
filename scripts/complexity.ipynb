```python
import pandas as pd
```

```python
df1 = pd.read_csv('results_TAASSC.csv')
df2 = pd.read_csv('L2SCA.csv')
df2.drop('writing_id', axis = 1, inplace = True)
df3 = pd.read_csv('results_TAALES.csv')
df3.drop('writing_id', axis = 1, inplace = True)
df4 = pd.read_csv('results_TAACO.csv')
df4.drop('writing_id', axis = 1, inplace = True)
df = df = pd.concat([df1,df2,df3,df4], axis = 1)
dtypes = ['str', 'float']
sentence = pd.read_excel('sentences.xlsx')
```

```python
# convert datetime to float
from datetime import datetime
bleu_score = []
for score in sentence.iloc[:, 81].values:
  if type(score) != str:
    s = f'{score.day}.{score.month}'
  bleu_score.append(float(s))

df['BLEU'] = bleu_score
bleu_score
```

```python
df
```

```python
text_names = []
for name in df.iloc[:, 0].values:
  pos = name.split('.')[0]
  text_names.append(int(pos))

bleu_score_sorted = []
for pos in text_names:
  bleu_score_sorted.append(bleu_score[pos-1])

df['BLEU'] = bleu_score_sorted
df
```
